```python
from typing import TypedDict

class PortfolioState(TypedDict):
    amount_usd: float
    total_usd: float
    total_inr: float
```

```python
def calc_total(state: PortfolioState) -> PortfolioState:
    state['total_usd'] = state['amount_usd'] * 1.0
    return state

def calc_total(state: PortfolioState) -> PortfolioState:
    state['total_inr'] = state['amount_usd'] * 85
    return state
```
